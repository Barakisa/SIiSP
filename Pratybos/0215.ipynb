```python
def check(x, w, b, f):
    res = True
    for xi in range(len(x[0])):
        if singleCheck([x[0][xi], x[1][xi]], w, b) != f[xi]:
            res = False
            break
    # if res:
    #     print("x: ", x, "w: ", w, "b: ", b)
    return res

def singleCheck(x, w, b):
    
    if w[0]*x[0] + w[1]*x[1] + b > 0:
        return 1
    else:
        return 0
```

```python
import matplotlib.pyplot as plt
import random

x = [[-0.2, 0.2, 0.8, 0.8], [0.5, -0.7, -0.8, 1]]
f = [0, 0, 1, 1]
w1static = 1

w2rand = random.uniform(-1, 1)
brand = random.uniform(-1, 1)
w = [w1static, w2rand]
res = [[],[]]
for i in range(100):
    k = 0
    while(check(x, w, brand, f) == False or k == 1000):
        k += 1  
        w2rand = random.uniform(-1, 1)  
        brand = random.uniform(-1, 1)
        w = [w1static, w2rand]
    res[0].append(w2rand)
    res[1].append(brand)

print("res: ", res)
plt.plot(res[0], res[1], 'ro')

plt.xlim(min(res[0]), max(res[0]))
plt.ylim(min(res[1]), max(res[1]))
plt.xlabel('w2')
plt.ylabel('b')
```
